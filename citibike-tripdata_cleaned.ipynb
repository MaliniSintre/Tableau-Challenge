```python
# Import dependencies
import pandas as pd
```

### Extract the Data

```python
# Read the data into Pandas DataFrames with explicit date format
Q1_2024 = pd.read_csv("Resources/citibike-tripdata-Q1-2024.csv", header=0, parse_dates=["started_at", "ended_at"], dayfirst=True)
```

```python
dtypes = {
    "ride_id": str,
    "rideable_type": str,
    "started_at": pd.Timestamp,
    "ended_at": pd.Timestamp,
    "start_station_name": str,
    "start_station_id": str,
    "end_station_name": str,
    "end_station_id": str,
    "start_lat": float,
    "start_lng": float,
    "end_lat": float,
    "end_lng": float,
    "member_casual": str
}
```

```python
Q1_2024.info()
```

```python
# Convert 'started_at' and 'ended_at' columns to datetime
Q1_2024['started_at'] = pd.to_datetime(Q1_2024['started_at'])
Q1_2024['ended_at'] = pd.to_datetime(Q1_2024['ended_at'])


# Separate date and time components
Q1_2024['start_date'] = Q1_2024['started_at'].dt.date
Q1_2024['start_time'] = Q1_2024['started_at'].dt.time
Q1_2024['end_date'] = Q1_2024['ended_at'].dt.date
Q1_2024['end_time'] = Q1_2024['ended_at'].dt.time

# Display the DataFrame with separated date and time components
print(Q1_2024[['start_date', 'start_time', 'end_date', 'end_time']])
```

```python
Q1_2024 = Q1_2024.rename(columns={
    "ride_id": "Ride ID",
    "rideable_type": "Rideable Type",
    "started_at": "Started At",
    "ended_at": "Ended At",
    "start_station_name": "Start Station Name",
    "start_station_id": "Start Station ID",
    "end_station_name": "End Station Name",
    "end_station_id": "End Station ID",
    "start_lat": "Start Latitude",
    "start_lng": "Start Longitude",
    "end_lat": "End Latitude",
    "end_lng": "End Longitude",
    "member_casual": "Member Type",
    "start_date": "Start Date",
    "start_time": "Start Time",
    "end_date": "End Date",
    "end_time": "End Time"

})

Q1_2024.head()
```

```python
Q1_2024.info()
```

```python
# Calculate duration
Duration = Q1_2024['Ended At'] - Q1_2024['Started At']

# Display the DataFrame with duration column
Duration
```

```python
Q1_2024['Duration'] = Duration
Q1_2024
```

```python
# Export the merged DataFrame to a CSV file
Q1_2024.to_csv("Resources/citibike-tripdata-Q1-2024_cleaned.csv", index=False)
```
